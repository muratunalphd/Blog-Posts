```python
import warnings
warnings.filterwarnings("ignore")
from plotnine import *
from tqdm import tqdm
from joblib import Parallel, delayed # for parallel processing
tqdm._instances.clear() 
from time import perf_counter
import statsmodels.formula.api as smf
```

```python
def dgp(**kwargs):
    # n, p =10000, 40 # no of obs. , features
    
    Xmat = np.random.multivariate_normal(np.zeros(p), np.eye(kwargs['p']), size=kwargs['n']).astype('float32')

    T = np.random.binomial(1, 0.5, n).astype('int8')

    col_list = ['X' + str(x) for x in range(1,(p+1))]

    df = pd.DataFrame(Xmat, columns = col_list)
    
    if kwargs['linear']:
        B = 10*df['X1'] + 5*df['X2'] + 20*(df['X3']-0.5) + 10*df['X4'] + 5*df['X5']
    else:
        B = 10*np.sin(np.pi*df['X1']*df['X2'] ) + 10*(df['X3']-0.5)**2 + 10*df['X4'] + 5*df['X5']
    
    # tau = df['X1'] + np.log(1 + np.exp(df['X2']))
    
    tau = 1
    
    Y = (B + tau*T + np.random.normal(0,25,1)).astype('float32')
    

    
    df['T'] = T
    df['Y'] = Y
    df['tau'] = tau
    
    return df
```

```python
%%time
import warnings
warnings.filterwarnings("ignore") 
from xgboost import XGBRegressor
from sklearn.model_selection import StratifiedKFold

def mlpredict(data):
    XGB_reg = XGBRegressor(learning_rate = 0.3, 
                            max_depth = 10, 
                            n_estimators = 500, 
                            reg_lambda = 1)

    kfold = StratifiedKFold(n_splits=2, shuffle=True, random_state=1)

    X_mat = data.columns.tolist()[0:40]

    ix = []
    Yhat = []
    for train_index, test_index in kfold.split(data, data["T"]):
        
        df_train = data.iloc[train_index].reset_index()
        df_test = data.iloc[test_index].reset_index()

        X_train = df_train[X_mat].copy()
        y_train = df_train['Y'].copy()
        X_test =  df_test[X_mat].copy()
        # y_test = df_test['Y'].copy()
        XGB_reg.fit(X_train, y_train)

        Y_hat = XGB_reg.predict(X_test)

        ix.extend(list(test_index))
        Yhat.extend(list(Y_hat))

    df_ml = pd.DataFrame({'ix':ix,'Yhat':Yhat}).sort_values(by='ix').reset_index(drop=True)
    df_ml[['Y','T']] = data[['Y','T']]
    df_ml['Ytilde'] = df_ml['Yhat'] - np.mean(df_ml['Yhat'])
    df_ml['Yres'] = df_ml['Y'] - df_ml['Yhat']
    df_ml = df_ml.drop('ix', axis=1)
    
    return df_ml
```

```python
def compute(linear=True):
    
    df = dgp(linear)
    
    #1. Difference-in-means
    mu_treated = np.mean(df.query('T==1')['Y'])
    mu_control = np.mean(df.query('T==0')['Y'])
    
    dim_tau = mu_treated - mu_control
    
    #2. OLS adjusted
    ols_adj = smf.ols('Y ~' + ('+').join(data.columns.tolist()[0:41]),
                 data = df).fit(cov_type='HC1',use_t=True)

    ols_adj_tau = ols_adj.params['T']
    
    #3. OLS interacted
    ols_int = smf.ols('Y ~' + 'T*('+('+').join(data.columns.tolist()[0:40])+')',
                 data = df).fit(cov_type='HC1',use_t=True)

    ols_int_tau = ols_int.params['T']
    
    #4. CUPED: OLS without 'T':
    df['Y_tilde'] = smf.ols('Y ~' + ('+').join(data.columns.tolist()[0:40]),
                 data = df).fit(cov_type='HC1',use_t=True).resid + np.mean(df['Y'])
    
    ols_tilde = smf.ols('Y_tilde ~ T', data=df).fit(cov_type='HC1',use_t=True)
    
    cuped_tau = ols_tilde.params['T']
    
    #5. Difference-in-differences
    pred_df = mlpredict(df)
    
    mu2_treated = np.mean(pred_df.query('T==1')['Yres'])
    mu2_control = np.mean(pred_df.query('T==0')['Yres'])
    
    did_tau = mu2_treated - mu2_control

    #6. MLRATE
    
    mlrate = ols = smf.ols('Y ~ T + Yhat + T:Ytilde',
                 data = pred_df).fit(cov_type='HC1',use_t=True)

    mlrate_tau = mlrate.params['T']
    
    return [dim_tau, ols_adj_tau, ols_int_tau, cuped_tau, did_tau, mlrate_tau]
```

```python
%%time
compute()
```

```python
%%time
compute(linear=False)
```

```python
%%time
tqdm._instances.clear() 
sim = 500

values = Parallel(n_jobs=8)(delayed(compute)() for _ in tqdm(range(sim)) )
results_df = pd.DataFrame(values, columns=['DIM', 'OLS_adj', 'OLS_int', 'CUPED','DID','MLRATE'])

for c in results_df.columns.tolist():
    print(f"Mean {c}: {round(results_df[c].mean(),3)}, 95% C.I.: {(round(np.percentile(results_df[c], 2.5),3), round(np.percentile(results_df[c], 97.5),3))}")
    
```

```python
data = dgp()
```

```python
data[['Y','T']].cov()
```

$\beta_1 = Cov(Y,x)/Var(x)$

```python
data[['Y','T']].cov()['Y']['T']/np.var(data['T'])
```

$\beta_k = Cov(Y,\tilde x_{k})/Var(\tilde x_k)$

> where $\tilde x_k$ is the residual from a regression of $x_k$ on all the other covariates

> each coef. in a multivariate regression is the bivariate slope coef. for the corresponding regressor after partialling out all the other covariates.

```python
data['model_t_res'] = smf.ols('T ~ ' + ('+').join(data.columns.tolist()[0:40]), data=data).fit(cov_type='HC1',use_t=True).resid
data[['Y','model_t_res']].cov()['Y']['model_t_res']/np.var(data['model_t_res'])
```

```python
 data['Tplot'] = data['T'] + np.random.normal(0, 0.01, size=len(data["Y"]))
(ggplot(data, aes(x='Tplot', y='Y'))+
 geom_point(color='c')+
 ylab("Spending Amount ($)") + 
 xlab("Received Coupon")
)
```

```python
print("Difference in means:",
      data.query("T==1")["Y"].mean() - data.query("T==0")["Y"].mean())

DIM = smf.ols('Y ~ T', data=data).fit(cov_type='HC1',use_t=True)
DIM.summary().tables[1]
```

```python
ols = smf.ols('Y ~' +('+').join(data.columns.tolist()[0:41]),data = data).fit(cov_type='HC1',use_t=True)

print(ols.params['T'])
print([ols.params['T'] - 1.96*ols.bse['T'] ] )
print( [ols.params['T'] + 1.96*ols.bse['T'] ])
```

```python
model_t = smf.ols('T ~ ' + ('+').join(data.columns.tolist()[0:40]), data=data).fit(cov_type='HC1',use_t=True)
model_y = smf.ols('Y ~ ' + ('+').join(data.columns.tolist()[0:40]), data=data).fit(cov_type='HC1',use_t=True)

residuals = pd.DataFrame(dict(res_y=model_y.resid, res_t=model_t.resid))

model_res = smf.ols('res_y ~ res_t', data=residuals).fit()
```

```python
print("Y Variance", np.var(data["Y"]))
print("Y Residual Variance", np.var(residuals["res_y"]))

print("T Variance", np.var(data["T"]))
print("T Residual Variance", np.var(residuals["res_t"]))

model_res.summary().tables[1]
```

```python
residuals['Tplot'] = data['Tplot']
(ggplot(residuals, aes(x='Tplot', y='res_y'))+
 geom_point(color='c') + 
 ylab("Spending Amount ($)") +
 xlab("Received Coupon")
)
```
